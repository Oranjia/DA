```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline  #设置图像显示方式
sns.set()    #设置画图空间为 Seaborn 默认风格
```

```python
df=pd.read_excel('DaPy_data.xlsx')
df.head()
```

```python
df["BMI"]=df["体重"]/((df["身高"]/100)*(df["身高"]/100))
```

```python
df.head()
# df["BMI"]
```

```python
def bmipy(number):
    if number<=18.4:
        return "偏瘦"   
    elif number>18.4 and number<=23.9:
        return "正常"
    elif number>23.9 and number<=27.9:
        return "过重"
    else:
        return "肥胖"
df["bmi"]=df["BMI"].map(bmipy)
df
```

```python
df["bmi"].values
```

```python

plt.rcParams['font.sans-serif']=['SimHei'] # 用来正常显示中文标签
#设置画布的大小
plt.figure(figsize=(10,10))
#第一个传入的是我们需要计算的数据，
plt.pie(df["bmi"].value_counts(),
       #labels是传入标签的
       labels=df["bmi"].values,    
       # 格式化输出百分比
       autopct='%.2f%%' 
       )
#绘图的标题
plt.title('班级的性别比例')
#关闭运行日志
plt.show()
```
