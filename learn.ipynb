```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline  #设置图像显示方式

```

```python
df=pd.read_excel(r'O:\VSCODE-DA\DaPy_data.xlsx')
df.head()
```

```python
"""统计身高分布"""
shengao=df["身高"].value_counts()
df_shengao=shengao.to_frame()
df_shengao["index"]=df_shengao.index
df_shengao.head()
```

```python
"""画身高分布图"""
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
fig, ax = plt.subplots(figsize=(15, 3))  # Create a figure containing a single axes.
ax.set_title("身高分布图")
ax.set_xlabel("身高")
ax.set_ylabel("频数")
ax.bar(df_shengao["index"],df_shengao["身高"],label='频数',width=0.01,color='g')
ax.legend()
ax.grid()
fig.show()
```

```python
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
fig, ax = plt.subplots(figsize=(15, 3))  # Create a figure containing a single axes.
ax.set_title("身高体重分布图")
ax.set_xlabel("样本")
ax.set_ylabel("身高体重")
ax.plot(df["体重"],'o',label='体重',color='g',linestyle='--')
ax.plot(df["身高"],'d',label='身高',color='r',linestyle='-')
ax.legend()
ax.grid()
fig.show()
```

```python
"""两个轴"""
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
fig, (ax,ax3) = plt.subplots(1,2,figsize=(15, 3))  # Create a figure containing a single axes.
ax.set_title("身高体重分布图")
ax.set_xlabel("样本")
ax.set_ylabel("体重")

ax2 = ax.twinx()
ax2.set_ylabel("身高")
l1,=ax.plot(df["体重"],'o',label='体重',color='g',linestyle='--')
l2,=ax2.plot(df["身高"],'d',label='身高',color='r',linestyle='-')
# ax.legend()
# ax2.legend()
ax.legend([l1,l2],["体重","身高"])
# ax.grid()

"""第二个图"""
# ...

fig.show()
```

```python
"""画个饼图"""
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
fig, ax = plt.subplots(1,2,figsize=(15, 3),facecolor="white",dpi=120)  # Create a figure containing a single axes.
# dpi=120 是像素密度
df['课程'].value_counts().plot.pie(ax=ax[0],shadow=False,autopct='%1.2f%%')
ax[0].set_ylabel('')    #设置y轴标签
ax[0].set_xlabel('课程')    #设置x轴标签
ax[0].legend(bbox_to_anchor=(1.0, 0.9))#设置图例的位置

ax[1].set_ylabel('')
# ax[1].set_xlabel('课程')
df['课程'].value_counts().head(5).plot.bar(ax=ax[1])  #前五个

# 显示数据标签
# 为每个条形图添加数值标签
for x,y in enumerate(df['课程'].value_counts()):
    ax[1].text(x, y+0.1, '%s' %round(y,1), ha='center')

fig.show()
```

```python
# 画个散点图
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
fig, ax = plt.subplots(figsize=(10, 3),facecolor="white",dpi=120)  # Create a figure containing a single axes.
# dpi=120 是像素密度
ax.set_xlabel("身高")
ax.set_ylabel("体重")
ax.scatter(
    df["身高"].values, 					# x轴坐标数组
    df["体重"].values,					# y轴坐标数组
    marker='o', 			# 点型
    s=100,				# 大小
    color='red',			# 颜色
    edgecolor='red', 		# 边缘颜色
    facecolor='green',		# 填充色
    zorder=1			# 图层序号
)
```
